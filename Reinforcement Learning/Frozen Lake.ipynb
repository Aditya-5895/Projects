```python
import gym
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output

# LEFT = 0 DOWN = 1 RIGHT = 2 UP = 3
# SFFF
# FHFH
# FFFH
# HFFG
import time
policy = {0: 2, 1: 1, 2: 1, 3: 0, 4: 1, 6: 1, 8: 2, 9: 1, 10: 1, 13: 2, 14: 2}

env = gym.make('FrozenLake-v0')
n_games = 1000
win_pct = []
scores = []

for i in range(3):
    done = False
    obs = env.reset()
    score = 0
    while not done:
        action = policy[obs]
        obs, reward, done, info = env.step(action)
#         clear_output(True)
        score += reward
        env.render()
        print(action)
        time.sleep(0.1)
    scores.append(reward)
print(scores)
```
