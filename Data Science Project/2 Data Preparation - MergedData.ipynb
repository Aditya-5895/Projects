## CurrentAttributesByZip

```python
CurrentAttributes = pd.read_excel(df,'CurrentAttributesByZip')
print(CurrentAttributes.shape)
CurrentAttributes.head()
```

```python
CurrentAttributes.columns = ['zip','sector']+list(CurrentAttributes.columns[2:])
CurrentAttributes.head()
```

```python
## Collecting the zipcode data matching with the labeled data
zip_attribute = pd.DataFrame()
for i in y_data['ZipCode']:
     zip_attribute = pd.concat([zip_attribute,CurrentAttributes[CurrentAttributes['zip']==i]])
print(zip_attribute.shape)
zip_attribute.head()
```

## RentalTrendsBySubmarket

```python
RentalTrendsBySubmarkets = pd.read_excel(df,"RentalTrendsBySubmarkets")
RentalTrendsBySubmarkets.head()
```

```python
RentalTrendsBySubmarkets = RentalTrendsBySubmarkets.dropna(axis=1,thresh=int(len(RentalTrendsBySubmarkets)*0.5))
RentalTrendsBySubmarkets.head()
```

## Mapping Zip_submarket with rental trends

```python
zip_to_submarket = pd.read_excel(df,'ZipToSubmarket')
zip_to_submarket.columns = ['zip','sector','market_name','submarket_name','number_of_properties']
zip_to_submarket.head()
```

```python
## Collecting the zipcode data matching with the labeled data
zip_submarket_data = pd.DataFrame()
for i in y_data['ZipCode']:
     zip_submarket_data = pd.concat([zip_submarket_data,zip_to_submarket[zip_to_submarket['zip']==i]])
print(zip_submarket_data.shape)
zip_submarket_data.head()
```

```python
zip_submarket_data = zip_submarket_data.merge(RentalTrendsBySubmarkets)
print(zip_submarket_data.shape)
zip_submarket_data.head()
```

## Rental Trends by market

```python
rental_trends_by_market  = pd.read_excel(df,"RentalTrendsByMarket")
print(rental_trends_by_market.shape)
rental_trends_by_market.head()
```

```python
rental_trends_by_market = rental_trends_by_market.dropna(axis=1,thresh=int(len(rental_trends_by_market)*0.5))
rental_trends_by_market.head()
```

## Mapping RentalTrends with ziptoMarket

```python
zip_to_market = pd.read_excel(df,'ZipToMarket')
zip_to_market.head()
```

```python
## Collecting the zipcode data matching with the labeled data
zip_market_data = pd.DataFrame()
for i in y_data['ZipCode']:
     zip_market_data = pd.concat([zip_market_data,zip_to_market[zip_to_market['zip']==i]])
zip_market_data.columns = ['zip','sector','market_name','number_of_properties']
print(zip_market_data.shape)
zip_market_data.head()
```

## Map

```python
zip_market_data = pd.merge(zip_market_data,rental_trends_by_market)
print(zip_market_data.shape)
zip_market_data.head()
```
