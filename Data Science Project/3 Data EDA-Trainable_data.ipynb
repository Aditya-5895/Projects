```python
for i in label_data:
    if label_data[i].nunique()<2:
        label_data = label_data.drop(i,axis=1)
print(label_data.shape)
label_data.head()
```

## Handeling missing value categorical and continous

```python
for i in label_data:
    if label_data[i].dtypes=='O':
        print(i)
        print(label_data[i].value_counts())
        print()
```

```python
label_data.shape,label_data.isnull().sum()
```

```python
median_dict = dict(label_data.median())
median_dict['submarket_name'] = 'Montrose/River Oaks'
median_dict['building_class'] = 'All'
median_dict['period'] = 'Y'
median_dict
```

```python
len(median_dict)
```

```python
label_data_new = label_data.fillna(value=median_dict,axis=0)
print(label_data_new.isnull().sum().sum())
label_data_new.head()
```

## Data Preprocessing
- Label Encoding
- Min Max Scaling
- Standardise

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
cleaned_labeled_data = label_data_new.copy()
categorical_dictonary = {}
for i in label_data_new.columns[1:-1]:
    if label_data_new[i].dtype =='O':
        le = LabelEncoder()
        cleaned_labeled_data[i] = le.fit_transform(label_data_new[i])
        categorical_dictonary[i] = dict(zip(le.classes_,list(range(len(le.classes_)))))
    else:
        min_max = StandardScaler()
        cleaned_labeled_data[i] = min_max.fit_transform(np.array(label_data_new[i]).reshape(-1,1))
```

## Removing corelated columns

```python
temp = cleaned_labeled_data.drop(['zip','Liking'],axis=1)
corr = abs(temp.corr())
corr.shape
```

```python
k = 0
for col in corr.columns[1:-1]:
    try:
        l = list(corr.columns[(corr.loc[col]>0.8) & (corr.loc[col]<1)])
        if len(l)!=0:
            corr = corr.drop(l,axis=1)
            corr = corr.drop(l,axis=0)
            print(f' {k+1} iteration removed columns are : {l}\n ')
            k+=1
    except:
        pass
```

```python
corr.shape
```
